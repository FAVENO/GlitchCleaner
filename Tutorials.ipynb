```python
import os
from GlitchCleaner import load_lora_parameters
from GlitchCleaner import count_passed_glitchtokens
# You may need to apply for a Hugging Face access token to use the model: https://huggingface.co/
# os.environ["HF_TOKEN"] = 'your_huggingface_token_here'
```

```python
# Load the repaired parameters
model_name = "Llama-2-7b-chat"  # Qwen-7B-Chat, Llama-2-7b-chat, gemma-2b-it, Mistral-7B-Instruct-v0.1,  Yi-6B-Chat, or Deepseek-llm-7b-chat
loaded_model = load_lora_parameters(model_name)
```

```python
# Effect of the repair
count_passed_glitchtokens(loaded_model, model_name)
```

```python
# Evaluate the model
from lm_eval import evaluator
from lm_eval.models.huggingface import HFLM

results = evaluator.simple_evaluate(
    model=HFLM(pretrained=loaded_model, tokenizer=loaded_model.tokenizer),
    tasks=["mmlu","gsm8k"], 
    verbosity="WARNING",
    confirm_run_unsafe_code=True,
)

print(results["results"])
```
